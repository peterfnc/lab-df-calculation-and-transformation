# Q2: How to identify which country has the most VIP Customers?

# Q3: How to identify which country has the most VIP+Preferred Customers combined?

Provide your solution for Q2 below:

```python
# your code here

Q95 = customers_df['amount_spent'].quantile(0.95)
Q75 = customers_df['amount_spent'].quantile(0.75)
```

```python
def check_customer_group(row):
    if row['amount_spent'] >= Q95:
        return 'VIP'
    elif row['amount_spent'] >= Q75:
        return 'Preferred'
    else:
        return 'Regular'
```

```python
customers_df['Category'] = customers_df.apply(lambda row: check_customer_group(row), axis = 1)

customers_df.head(10)
```

```python
orders_customer = pd.merge(orders, customers_df, on ='CustomerID', how= 'left')

orders_customer.head(10)
```

```python
columns = ['CustomerID', 'Country', 'Category']
country_customer = pd.DataFrame(list(orders_customer.groupby(columns).groups.keys()), columns = columns)

country_customer.head()
```

```python
country_vips = country_customer[country_customer['Category'] == 'VIP']
country_vips.head()
```

```python
country_count = country_vips[['Country', 'CustomerID']].groupby('Country').count()
country_count.head(100)
```

```python
country_count.nlargest(1, 'CustomerID')
```

```python
country_vip_pref = country_customer[country_customer['Category'].isin(['VIP','Preferred'])]
country_vip_pref.head()
```

```python
country_count = country_vip_pref[['Country', 'CustomerID']].groupby('Country').count()
country_count.head(100)
```

```python
country_count.nlargest(1, 'CustomerID')
```
